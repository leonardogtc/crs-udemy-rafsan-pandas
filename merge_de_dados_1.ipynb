```python
import pandas as pd
```

```python
# Cadastro da Loja A
cadastro_a = {
    'Id':['AA2930','BB4563','CC2139','DE2521','GT3462','HH1158'],
    'Nome':['Victor','Amanda','Bruna','Carlos','Ricardo','Maria'],
    'Idade':[20,35,40,54,30,27],
    'CEP':['20365-777','20735-050','25896-896','45699-889','12345-698','45889-569']
}
cadastro_a = pd.DataFrame(cadastro_a, columns = ['Id','Nome','Idade','CEP'])
cadastro_a
```

```python
# Cadastro da Loja b
cadastro_b = {
    'Id':['SS4567','DF4567','CC2339','BB4563','GT3462','HH1158'],
    'Nome':['Jonas','Clara','Lívia','Amanda','Ricardo','Maria'],
    'Idade':[22,21,40,35,30,27],
    'CEP':['20365-767','20735-040','25896-888','20735-050','12345-698','45889-569']
}
cadastro_b = pd.DataFrame(cadastro_b, columns = ['Id','Nome','Idade','CEP'])
cadastro_b
```

```python
# Cadastro da Loja c
cadastro_c = {
    'Id':['AA2930','BB4563','CC2139','AA2930','GY6789','HH1158'],
    'Nome':['Victor','Amanda','Bruna','Victor','Ricardo','Maria'],
    'Idade':[20,35,40,54,30,27],
    'CEP':['20365-777','20735-050','25896-896','20365-777','12345-678','45889-569']
}
cadastro_c = pd.DataFrame(cadastro_c, columns = ['Id','Nome','Idade','CEP'])
cadastro_c
```

```python
# Registro de compras
compras = {
    'Id':['AA2930','CC2139','HH1158','GT3462','DF4567','BB4563'],
    'Data':['2023-11-01','2023-11-13','2023-11-13','2023-11-14','2023-11-15','2023-11-15'],
    'Valor':[600,555,345,890,1024,2048]
}
compras = pd.DataFrame(compras, columns=['Id','Data','Valor'])
compras
```

```python
# Inner Join
# É como no SQL - Mostra o que há de comum nos dois.
pd.merge(cadastro_a, cadastro_b, on=["Id"], how="inner")
```

```python
# Inner Join
# Alguns campos
pd.merge(cadastro_a, cadastro_b[['Id','Idade','CEP']], on=['Id'], how='inner')
```

```python
# Inner Join
# Alguns campos
pd.merge(cadastro_a, cadastro_b[['Id','Idade','CEP']], on=['Id'], how='inner', suffixes=('_A','_B'))
```

Full Join
---------

```python
lojas = pd.concat([cadastro_a, cadastro_b], ignore_index=True)
```

```python
# Remover duplicadas da loja
clientes_unicos = lojas.drop_duplicates(subset='Id')
```

```python
clientes_unicos
```

```python
# clientes que fizeram compras
# e pertencem ao cadastro da loja A
pd.merge(cadastro_a, compras, on=['Id'], how='left')
```

Outer Join

```python
pd.merge(cadastro_a, cadastro_b, how="outer", on=['Id'], indicator=True)
```
