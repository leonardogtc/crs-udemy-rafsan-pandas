```python
import pandas as pd
import numpy as np
```

```python
# Range de dias
dias = pd.date_range("20230101", periods=12)
```

```python
dias
```

```python
pessoas = ['George', 'Victor', 'Lucas']
```

```python
# Escolher um item dentro da lista
np.random.choice(pessoas)
```

```python
Nome = []
Gasto = []
for i in range(12):
  Nome.append(np.random.choice(pessoas))
  Gasto.append( np.round(np.random.rand()*100, 2) )
Nome
```

```python
Gasto
```

```python
# Criando o DataFrame
df = pd.DataFrame({'Dia':dias, 'Nome':Nome, 'Gasto':Gasto })
```

```python
df
```

```python
# Agrupar dados
df.pivot(index='Dia', columns='Nome', values='Gasto')
```
