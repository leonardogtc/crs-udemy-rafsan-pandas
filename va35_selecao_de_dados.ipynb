```python
import pandas as pd
import numpy as np
```

```python
# Criar date_range
datas = pd.date_range('20180101', periods=600, freq="D")
```

```python
# Criar dataFrame
df = pd.DataFrame(np.random.randn(600, 5), index= datas, columns=list("ABCDE"))
```

```python
df
```

```python
# Obtendo dados somente de uma coluna
df['D'].head(5)
```

```python
# Obtendo linha de 2 a 5
df[1:5]
```

```python
# Obtendo todas as linhas das colunas B, C e D
df.loc[:,['B', 'C', 'D']]
```

```python
# Obtendo a filtragem de linhas pelas datas do DataFrame
f = df.loc['2018-03-01':'2018-03-31',['A','E']]
```

```python
# NÃºmero de linhas da filtragem
len(f)
```

```python
# Coletando os dados de uma linha especÃ­fica
df.iloc[1]
```

```python
# Coletando os dados de uma linha e colunas especÃ­ficas
# Linha 1 atÃ© linha 4 nÃ£o inclusa e colunas de Ã­ndice 0 atÃ© Ã­ndice 2 nÃ£o incluso
df.iloc[1:4, 0:2]
```

```python
# Coletando linhas e colunas por Ã­ndices especificados nÃ£o sequenciais
df.iloc[[0,2,4,6,8],[1,2]]
```

```python
# Coletar linhas especÃ­ficas definidas e todas as colunas
df.iloc[1:3, :]
```

```python
# AULA 37 - Filtros Booleanos
# Obter todos os valores de A que forem maiores que zero
df[df.A > 0]
```
